```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei'] #显示中文字体
plt.rcParams['axes.unicode_minus'] = False # 显示中文负号
```

```python
df = pd.read_excel('../data/ftx.xls')
df.head()
```

查询不同区域销售数量均值  折线图

```python
df.groupby('销售区域').mean()['销售数量'].plot()
```

柱状图

```python
df.groupby('销售区域').mean()['销售数量'].plot(kind='bar')
```

```python
df.groupby('销售区域').mean()['销售数量'].plot(kind='pie', autopct='%1.1f%%')
```

```python
df.plot(kind='scatter', x='年', y='月')
```

可以同时显示多个折线图

```python
df.plot(x='年', y=['月', '销售数量'])
```

使用bank.xls完成以下可视化效果。。。  
使用条形图统计还款状态的个数

```python
df = pd.read_excel('../data/bank.xls')
df.head()
```

```python
df['还款状态'].value_counts().plot(kind='bar')
```

使用折线图显示数据中贷款最高的前100项数据值

```python
df.sort_values('贷款金额', ascending=False).head(100)['贷款金额'].reset_index().plot(y='贷款金额')
```

使用饼状图显示不同贷款期限的所占的比例

```python
df['贷款期限'].value_counts().plot(kind='pie', autopct='%1.1f%%', shadow=0.2)
```
