数据表合并处理  
交叉表

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.DataFrame({'姓名':['tom', 'alex', 'john'], '年龄':[20, 25, 30]})
df1
```

```python
df2 = pd.DataFrame({'姓名':['tom', 'alex', 'adam'], 
                    '爱好':['football', 'smoking', 'playgame']})
df2
```

df1, df2都有姓名列，姓名有相同项， 合并两个数据

```python
df = pd.merge(df1, df2) # 合并数据
df
```

```python
help(pd.merge) # 查看函数帮助
```

左连接 how='left'  
应用场景：多张表格存储不同数据

```python
df = pd.merge(df1, df2, on='姓名', how='left')# 两张表使用姓名列关联， 保留左表全部数据
df
```
