逻辑回归  
二分类    扩展多分类（OVO，ovr）
与线性回归原理类似（分类线 $ z=wx+b, sigmoid函数，用来计算样本正类别概率 \hat{y}=\dfrac{1}{1+e^{-z}} $）

```python
# 调库
from sklearn.datasets import load_iris # 鸢尾花数据集   150条  4个特征  3个分类
from sklearn.linear_model import LogisticRegression # 逻辑回归 分类模型
from sklearn.model_selection import train_test_split, GridSearchCV # 数据切分， 模型调参
from sklearn.preprocessing import StandardScaler, OneHotEncoder # 数据预处理操作
import warnings
warnings.filterwarnings('ignore')
```

```python
# 提取数据
x, y = load_iris(return_X_y=True)
print(x.shape)
print(y.shape)
print(x[:5])
```

```python
# 特征都是连续值，需要特征缩放
std = StandardScaler()
# 转化x进行标准化处理
x = std.fit_transform(x)
# y一维数据，需要转换为二维数据
y = y.reshape(-1, 1)
print(x[:5])
print(y.shape)
```

```python
# 切分数据集
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
```

```python
help(LogisticRegression)
```

```python
# 逻辑回归调参处理
lr = LogisticRegression() # 创建模型
pg = {'penalty': ['l1', 'l2'], 'C': [1,2,5,10]}# penalty 选择处理的正则化方式  C 正则化系数的倒数
model = GridSearchCV(lr, pg, cv=3)
model.fit(x_train, y_train)
print(model.best_score_)
print(model.best_params_)
```

```python
# 选择最优参数创建模型
model = LogisticRegression(penalty='l1', C=5)
model.fit(x_train, y_train)
print('测试集得分：', model.score(x_test, y_test))
# 每一行代表的是一个样本对应三个类别预测的概率， 三个概率和为1，预测的类别是每一行中概率最大值
print('测试集的概率\n', model.predict_proba(x_test)[:5])
print('测试集预测结果\n', model.predict(x_test)[:5])
```
