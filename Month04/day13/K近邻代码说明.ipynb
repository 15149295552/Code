```python
# 调库
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor 
```

n_neighbors: K值  
weights : 是否使用带权处理   默认：uniform 不带权  distance： 带权  （距离的倒数）  
p： 默认2，欧式距离   1曼哈顿距离

```python
# 使用鸢尾花数据，使用KNN算法调参，打印模型最终准确率
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.datasets import load_iris
# 训练集：测试集 7:3进行切分  设置随机种子为234 ，最优参数是多少，最后准确率
```

```python
# 加载数据
x, y = load_iris(return_X_y=True)
x[:5]
```

```python
# 特征缩放    归一化、标准化
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
x = StandardScaler().fit_transform(x) # 数据转化
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=234)
```

```python
# 调参
knn = KNeighborsClassifier()
pg = {'n_neighbors': [3, 4, 5, 6, 7], 'weights': ['uniform', 'distance'], 'p': [1, 2]}
model = GridSearchCV(knn, pg, cv=5)
model.fit(x_train, y_train)
model.best_params_
```

```python
# 使用计算得到的最优参数创建模型，查看测试集得分
knn = KNeighborsClassifier(n_neighbors=5, p=2, weights='distance')
knn.fit(x_train, y_train)
knn.score(x_test, y_test) # 最后得分
```
